```python
# преобразование в категории тех столбцов, в которых меньше 150 уникальных значений
melb_df.nunique() # количество уникальных значений в каждом столбце
to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car'] # за исключением этих значений
for column in melb_df.columns:
    if melb_df[column].nunique() < 150 and column not in to_exclude:
        melb_df[column] = melb_df[column].astype('category') # преобразование типа
melb_df.info()
```
