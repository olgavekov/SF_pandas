#### DataFrame. Create copy. Display first 5 raws

```python
melb_df = melb_data.copy()
melb_df.head()
```

#### Delete columns from DataFrame

```python
melb_df = melb_df.drop(['index', 'Coordinates'], axis=1)
melb_df.head()
```

#### Create new column in DataFrame

```python
melb_df['total_rooms'] = melb_df['Rooms'] + melb_df['Bedroom'] + melb_df['Bathroom']
display(melb_df['total_rooms'])
```

```python
melb_df['MeanRoomsSquare'] = melb_df['BuildingArea'] / melb_df['total_rooms']
display(melb_df['MeanRoomsSquare'])
```

#### New statistical measures

```python
diff_area = melb_df['BuildingArea'] - melb_df['Landsize']
sum_area = melb_df['BuildingArea'] + melb_df['Landsize']
melb_df['AreaRatio'] = diff_area/sum_area
display(melb_df['AreaRatio'])
```

```python
countries_df = pd.DataFrame({
    'country': ['Англия', 'Канада', 'США', 'Россия', 'Украина', 'Беларусь', 'Казахстан'],
    'population': [56.29, 38.05, 322.28, 146.24, 45.5, 9.5, 17.04],
    'square': [133396, 9984670, 9826630, 17125191, 603628, 207600, 2724902]
})
countries_df['density'] = countries_df['population'] * 1000000 / countries_df['square']
countries_df.head(7)
countries_df['density'].mean()
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'], dayfirst=True)
display(melb_df['Date'])
```

```python
years_sold = melb_df['Date'].dt.year
print(years_sold)
print('Min year sold:', years_sold.min())
print('Max year sold:', years_sold.max())
print('Mode year sold:', years_sold.mode()[0])
```

```python
unique_list = []                                            # создаём пустой список
for col in melb_df.columns:                                 # пробегаемся по именам столбцов в таблице
    item = (col, melb_df[col].nunique(),melb_df[col].dtype) # создаём кортеж (имя столбца, число уникальных значений)
    unique_list.append(item)                                # добавляем кортеж в список
unique_counts = pd.DataFrame(                               # создаём вспомогательную таблицу и сортируем её
    unique_list,
    columns=['Column_Name', 'Num_Unique', 'Type']
).sort_values(by='Num_Unique',  ignore_index=True)
display(unique_counts)                                      # выводим её на экран
```
